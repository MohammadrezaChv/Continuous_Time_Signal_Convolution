```python
import numpy as np
import matplotlib.pyplot as plt

def convolve1D(signal1, signal2):
    # Get the lengths of the signals
    signal1_len = len(signal1)
    signal2_len = len(signal2)
    
    # Compute the length of the output
    output_len = signal1_len + signal2_len - 1
    
    # Create an empty output array
    result = np.zeros(output_len)
    
    # Perform the convolution
    for i in range(output_len):
        for j in range(signal2_len):
            if i - j >= 0 and i - j < signal1_len:
                result[i] += signal1[i - j] * signal2[j]
    
    return result

# Define the 1D signals
signal1 = np.array([1, 2, 3, 4, 5])
signal2 = np.array([1, 0, -1])

# Perform the convolution
result = convolve1D(signal1, signal2)

# Plotting the signals
plt.figure(figsize=(8, 6))

# Plot the first input signal
plt.subplot(311)
plt.plot(signal1)
plt.title('Input Signal 1')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)

# Plot the second input signal
plt.subplot(312)
plt.plot(signal2)
plt.title('Input Signal 2')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)

# Plot the output signal
plt.subplot(313)
plt.plot(result)
plt.title('Convolution Result')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)

plt.tight_layout()
plt.show()
```
